# pklファイルをjson形式に変換する

### ライブラリのインポート

```python
import pickle
import json
import numpy as np
```

### 定数設定

```python
# パスの設定
pkl_path  = "../data_nuscenes/nuscenes/bevdetv3-nuscenes_infos_val.pkl"
json_path = "../tools/bevdetv3-nuscenes_infos_val.json"

# 出力するkeyの設定
target_keys = ["token", "scene_token", "timestamp", 
               "lidar2ego_translation", "lidar2ego_rotation", 
               "ego2global_translation", "ego2global_rotation",
               "cams"]
```

### jsonにdumpするする

```python
def to_jsonable(obj):
    """dumpできる形式に変換する"""
    if isinstance(obj, dict):
        return {str(k): to_jsonable(v) for k, v in obj.items()}
    elif isinstance(obj, (list, tuple, set)):
        return [to_jsonable(v) for v in obj]
    elif isinstance(obj, np.ndarray):
        return obj.tolist()
    elif isinstance(obj, np.integer):
        return int(obj)
    elif isinstance(obj, np.floating):
        return float(obj)
    elif isinstance(obj, np.bool_):
        return bool(obj)
    elif obj is None or isinstance(obj, (str, int, float, bool)):
        return obj
    else:
        return str(obj)
        
# pklファイルの読み込み
with open(pkl_path, "rb") as f:
    data = pickle.load(f)

# infosから必要なデータを選択
selected_infos = []

for info in data["infos"]:
    item = {}

    for k in target_keys:
        if k in info:
            item[k] = info[k]

    selected_infos.append(item)

# dump用の辞書を作成
out = {
    "metadata": data["metadata"],
    "infos": selected_infos
}

# jsonにdump
with open(json_path, "w", encoding="utf-8") as f:
    json.dump(to_jsonable(out), f, ensure_ascii=False, indent=2)
    
print(f"saved: {json_path}")
```
